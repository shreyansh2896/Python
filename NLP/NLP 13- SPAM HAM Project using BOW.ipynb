```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word  in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

###  BOW 

```python
## Output Features

y=pd.get_dummies(messages['label'])
y=y.iloc[:,0].values   # We can pick either of the column so we have taken only 1 column ham
```

```python
## Train test split 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(corpus,y,test_size=0.2)
```

```python
# Create the bag of words model

from sklearn.feature_extraction.text import CountVectorizer

# For binary BOW enable binary = True

cv=CountVectorizer(max_features=3000,ngram_range=(1, 2))
```

```python
x_train
```

```python
## Independent Features

x_train=cv.fit_transform(x_train).toarray()
```

```python
x_train    # Converted sentences into vectors
```

```python
len(x_train),len(y_train)
```

```python
x_test   # X-test is a sentence and we need to convert this into vectors
```

```python
x_test=cv.transform(x_test).toarray()   # We are not fitting the data into the model, simply transforming it as using fit it is going to learn from the data.
```

```python
x_test
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
x_train
```

```python
cv.vocabulary_
```

```python
## Use of Machine learning algorithm

from sklearn.naive_bayes import MultinomialNB   # NB is used as it is very good in handling the sparse matrix
```

```python
spam_detect_model=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
## Comparision of y_test and y_pred

from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
CR=classification_report(y_test,y_pred)
print(CR)
```
